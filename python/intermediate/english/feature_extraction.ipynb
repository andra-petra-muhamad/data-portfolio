# Feature Extraction
Feature extraction is the process of generating new features from existing features to enrich the information available to a machine learning model. The goal is to capture hidden patterns or relationships between variables that may not be immediately apparent from the raw data. Some common techniques in feature extraction include:
* Calculating ratios between numeric features
* Logical or mathematical transformations of features
* Merging or subtracting between columns
* Modifying and recoding categorical variables

This feature extraction process provides new, more meaningful and contextual features, which can help improve the overall performance and interpretability of a predictive model.

## Loading the California Housing Dataset
Loading the California Housing Train dataset from a local CSV file into a DataFrame for exploration and feature extraction.

```python
import pandas as pd
```

```python
dataset = pd.read_csv("C:/Users/LENOVO/Python/california_housing_train.csv")
```

```python
dataset.head()
```

## Calculating the Number of Rooms Other Than Bedrooms
Calculating the number of rooms other than bedrooms to determine the proportion of other spaces in a house.

```python
other_rooms = dataset["total_rooms"] - dataset["total_bedrooms"]
other_rooms
```

## Calculating Population Density
Calculating population density is based on the ratio of the population to the number of households.

```python
population_density = dataset["population"].div(dataset["households"])
density_alt = dataset["population"]/dataset["households"]
density_alt
```

## Adding Density Feature to Dataset
Adds a new feature called density to the dataset as a result of the previous density calculation.

```python
dataset["population_density"] = population_density
dataset
```

## Alternative Ways to Add a Density Feature
Another way to add a density column to a dataset is to use the assign() function.

```python
dataset = dataset.assign(population_density = population_density)
dataset
```

## Creating the Price to Income Ratio Feature
Insert a new feature, price_income_ratio, in the third column, which represents the ratio between house prices and median income.

```python
dataset.insert(2, "price_income_ratio", dataset["median_house_value"] / dataset["median_income"])
dataset
```

## Calculating Room Density
Measuring the average number of rooms per household to see the room density per dwelling.

```python
room_density = dataset["total_rooms"] / dataset["households"]
room_density
```

## Creating a Quadratic Space Density Feature
Creates a new feature that squares the spatial density to model possible non-linear relationships.

```python
dataset.assign(room_density_squared = room_density.pow(2))
```

## Loading the MPG Dataset from Seaborn
Loading the MPG (Miles Per Gallon) dataset from Seaborn for feature extraction on vehicle data.

```python
import seaborn as sns
```

```python
mpg_dataset = sns.load_dataset("mpg")
```

```python
sns.get_dataset_names()
```

```python
mpg_dataset["origin"].unique()
```

## Standardizing Origin Labels to America
Replacing the country of origin label from "usa" to "america" to standardize category naming.

```python
mpg_dataset.loc[mpg_dataset["origin"] == "usa", "origin"] = "america"
mpg_dataset
```

## Adding a New Column for Country Labels
Creates a new column, new_origin, that represents the results of changing the vehicle's origin name, without replacing the original column.

```python
mpg_dataset.loc[mpg_dataset["origin"] == "usa", "new_origin"] = "america"
mpg_dataset
```

## Transforming Origin Labels to New Origin Labels
Performed a gradual category transformation by replacing the label "japan" with "asia" to simplify regional analysis.

```python
mpg_dataset["new_origin"] = mpg_dataset["origin"].replace("japan", "asia")
mpg_dataset.query("origin == 'asia'")
```

## Remapping Region Labels to Country of Origin
Reverts the vehicle's origin category to its original form using dictionary mapping for label standardization.

```python
mpg_dataset["new_origin"] = mpg_dataset["origin"].map({"america": "usa",
                                                      "asia" : "japan",
                                                      "europe" : "europe"
})
mpg_dataset
```

## Renaming Columns to Make Them More Descriptive
Renaming columns to make them more descriptive and easier to understand, such as changing "origin" to "country."

```python
mpg_dataset = mpg_dataset.rename(columns = {"origin": "country",'mpg':'miles per galon'})
mpg_dataset
```

## Removing Irrelevant Columns
Removing columns deemed irrelevant or unnecessary for further analysis.

```python
mpg_dataset = mpg_dataset.drop(columns = ['miles per galon','model_year'])
mpg_dataset
```

## Deleting Temporary Columns
Deletes the new_origin column previously used for temporary label manipulation.

```python
mpg_dataset.drop(columns = "new_origin", inplace = True)
```

```python
mpg_dataset
```

Filters vehicle data originating from Japan to focus on analyzing a subset of the data based on country of origin.
