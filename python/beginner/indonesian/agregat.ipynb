### Customize

over_value digunakan untuk mengetahui seberapa banyak tipe area dan penyewa yang melebihi ambang batas

```python
def over_value(x, nilai = 500):
  return sum(x > nilai)
```

```python
by_area_tenant["Rent"].agg(["min", "max", "median", "mean", over_value])
```

```python
by_area_tenant.agg({"Rent" : ["min", "max", "mean"],
                    "Size" : ["min", "max"]})
```

## Transform

Membuat kolom median dari harga sewa

```python
house_data["median_rent_by_area"] = rent_by_area["Rent"].transform("median")
```

Hanya menampilkan baris dengan harga sewa di atas median

```python
house_data[house_data["Rent"] >= house_data["median_rent_by_area"]]
```

## Rata-Rata Sewa

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Hitung rata-rata sewa per area
avg_rent = house_data.groupby("Area Type")["Rent"].mean().sort_values()

# Plot
avg_rent.plot(kind='barh', title="Rata-Rata Sewa per Area")
plt.xlabel("Rent (INR)")
plt.show()
```

## Ranking Lokal

rank() adalah peringkat (urutan) numerik yang diberikan ke setiap nilai dalam suatu kolom, biasanya berdasarkan besar-kecilnya nilai.

Catatan: misal 5000 muncul dua kali â†’ mereka berbagi peringkat 1 dan 2 â†’ (1+2)/2 = 1.5

```python
house_data["rank_rent_in_area"] = house_data.groupby("Area Type")["Rent"].rank(ascending=False)
#house_data["rank_rent_in_area"] = house_data.groupby(["Area Type","Area Locality"])["Rent"].rank(ascending=False)
```

Rumah itu berada di peringkat ke-1622 dari seluruh rumah di area-nya, berdasarkan harga sewa (dari yang paling tinggi ke paling rendah).
 
Rumah dengan sewa 5 teratas di setiap Area Type

```python
house_data[house_data["rank_rent_in_area"] <= 5]
```
