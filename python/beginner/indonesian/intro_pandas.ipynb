# Pandas Series

```python
import pandas as pd

a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)
```

```python
print(myvar[0])
```

```python
import pandas as pd

a = [1, 7, 2]

myvar = pd.Series(a, index = ["x", "y", "z"])

print(myvar)
```

Setelah membuat label, Kita dapat mengakses item dengan merujuk ke label.

```python
print(myvar["y"])
```

## Key/Value Objects as Series

> Indented block



```python
import pandas as pd

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)
```

Untuk memilih hanya beberapa item dalam kamus, gunakan indeks argumen dan hanya tentukan item yang ingin Kita sertakan dalam Seri.

```python
import pandas as pd

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories, index = ["day1", "day2"])

print(myvar)
```

## DataFrame

```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

myvar = pd.DataFrame(data)
df = pd.DataFrame(data)
print(myvar)
print(df)
```

# Locate Row

```python
# merujuk ke indeks baris:
print(df.loc[0])
```

```python
# menggunakan list dari indeks:
print(df.loc[[0, 1]])
```

## Named Index

```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])

print(df)
```

### Locate Named Indexes

```python
df.loc["day1"]
```

```python
# merujuk ke indeks yang ternama:
print(df.loc["day3"])
```
