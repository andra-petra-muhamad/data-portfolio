```python
dataset = pd.read_csv("C:/Users/LENOVO/Python/california_housing_test.csv")
```

Ambil data dengan kriteria usia median perumahan dan populasi

```python
dataset[(dataset["housing_median_age"] < 20) &
        dataset["population"] < 100]
```

* items=[1,4]: ambil label index 1 dan 4
* axis=0: berarti ambil baris
* Perlu diingat bahwa filter tidak bisa menggunakan logika dan hanya berguna ketika tau nama kolom atau posisi data

```python
print(dataset.filter(items=[1,4],axis=0))
```

menggunakan query seperti MySQL untuk mengambil data dengan logika

```python
dataset.query("housing_median_age < 20 and population < 100")
```

```python
print(dataset["population"] < 100)
```

```python
dataset.sort_values(by = "housing_median_age", ascending = False)
```

* Mengurutkan data berdasarkan housing_median_age secara menurun (descending, dari besar ke kecil)
* Jika nilai housing_median_age sama, maka baris-baris data tersebut akan diurutkan berdasarkan population secara menaik (ascending, dari kecil ke besar)

```python
dataset.sort_values(by = ["housing_median_age", "population"],
                    ascending = [False, True])
```
