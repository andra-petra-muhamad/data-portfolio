# Seaborn

## Scatterplot

Scatter plots are plotted using the scatterplot() method. This is similar to Matplotlib, but additional argument data is required.

```python
sns.scatterplot(x='day', y='tip', data=data,)
plt.show()
```

We'll find that using Matplotlib it can be quite difficult if we want to color each plot point according to the sex column. However, in a scatter plot, this can be done with the hue argument.

# Customize plot in Seaborn

## Change the aesthetics of the figure

set_style() is used to set the aesthetics of a plot. This means it affects things like axis colors, whether the grid is active or not, and other aesthetic elements. There are five themes available in Seaborn:

- darkgrid
- whitegrid
- dark
- white
- tick

```python
# loading dataset
data = sns.load_dataset("iris")

sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Change the theme to the dark
sns.set_style("darkgrid")
plt.show()
```

## Spine removal

The spine is a line that marks the boundaries of the data and connects the axis tick marks. It can be removed using the despine() method.

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Delete spine
sns.despine()
plt.show()
```

# Changing Figure Size
Figure sizes can be changed using Matplotlib's figure() method. The figure() method creates a new figure with the specified size passed in the figsize parameter.

```python
# Change the size of the figure
plt.figure(figsize = (2, 4))

sns.lineplot(x="sepal_length", y="sepal_width", data=data)
sns.despine()
plt.show()
```

## Plot Scaling

This is done using the set_context() method. This allows us to override the default parameters. This affects things like the size of labels, lines, and other plot elements, but not the overall style. The default context is "notebook," and the other contexts are "paper," "talk," and "poster." font_scale sets the font size.

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Adjust the plot scale
sns.set_context("paper")
plt.show()
```

## Setting Temporary Styles
The axes_style() method is used to set temporary styles. It is used in conjunction with the with statement.

```python
def plot():
	sns.lineplot(x="sepal_length", y="sepal_width", data=data)

with sns.axes_style('darkgrid'):
    
	# Add subplots
	plt.subplot(211)
	plot()
plt.subplot(212)
plot()
```

# Color Palette

```python
# the color of the palette now
palette = sns.color_palette()

# pallet color plot as a horizontal array
sns.palplot(palette)

plt.show()
```

## Divergent Color Palette
This type of color palette uses two different colors, each representing a different point starting from the same point in both directions. Consider a range of -10 to 10, so that values from -10 to 0 represent one color, and values from 0 to 10 represent the other color.

```python
# the color of the palette now
palette = sns.color_palette('PiYG', 10)

# divergent color palette
sns.palplot(palette)

plt.show()
```

In the example above, we've used the built-in divergent color palette, which displays 11 different color points. The color on the left represents pink, and the color on the right represents green.

## Sequential Color Palette
A sequential palette is used where the distribution ranges from lower to higher values. To do this, add an 's' to the color passed in the color palette.

```python
# the color of the palette now
palette = sns.color_palette('Blues', 11)

# ranks of the color of the palette
sns.palplot(palette)

plt.show()
```

## Setting the Default Color Palette
The set_palette() method is used to set the default color palette for all plots. The arguments for color_palette() and set_palette() are the same. set_palette() changes the default matplotlib parameters.

```python
data = sns.load_dataset("iris")

def plot():
	sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Set the default color of the palette
sns.set_palette('vlag')
plt.subplot(211)

# plot with vlag palette color
plot()

# Set the default color of the other palette
sns.set_palette('Accent')
plt.subplot(212)
plot()

plt.show()
```

# Multipleplots Seaborn

## Method 1: Using the FacetGrid() method

- The FacetGrid class helps visualize the distribution of a single variable as well as the relationships between multiple variables separately within subsets of your data set using multiple panels.
- A FacetGrid can be drawn in up to three dimensions: row, col, and hue. The first two have a clear correspondence with the resulting axis layout; think of the hue variable as a third dimension along the depth axis, where different levels are plotted with different colors.
- The FacetGrid object takes as input a dataframe and the names of the variables that will form the row, column, or hue dimensions of the grid. The variables must be categorical, and the data at each level of the variable will be used for the facets along that axis.

```python
data = sns.load_dataset("iris")

plot = sns.FacetGrid(data, col="species")
plot.map(plt.plot, "sepal_width")

plt.show()
```

## Method 2: Using the PairGrid() method
- A grid subplot to plot pairwise relationships in a data set.
- This class maps each variable in the data set to a column and row in a multi-axis grid. Different axis-level plotting functions can be used to draw bivariate plots in the upper and lower triangles, and the marginal distributions of each variable can be shown on the diagonal.
- It can also represent an additional level of conventionalization with the hue parameter, which plots different subsets of the data in different colors. This uses color to resolve elements in the third dimension, but only plots subsets on top of each other and will not adjust the hue parameter for a specific visualization as the axis-level functions that accept hue do.

```python
# loading dataset
data = sns.load_dataset("flights")
plot = sns.PairGrid(data)
plot.map(plt.plot)

plt.show()
```

# Make various types of plots


## Relational Plots
Relational plots are used to visualize statistical relationships between data points. Visualization is necessary because it allows humans to see trends and patterns in data. The process of understanding how variables in a data set relate to each other and their relationships is called statistical analysis. See the article below for detailed information.

__relplot()__

This function gives us access to several other axis-level functions that show the relationship between two variables using semantic subset mapping. It is plotted using the relplot() method.

```python
# loading dataset
data = sns.load_dataset("iris")

# make a reclot
sns.relplot(x='sepal_width', y='species', data=data)

plt.show()
```

There are also scatterplots and lineplots

## Categorical Plot
Categorical plots are used when we need to visualize the relationship between two numerical values. A more specialized approach can be used if one of the primary variables is categorical, meaning it takes on a fixed and limited number of possible values.

One example of the categorical is Barplot

then there is also a boxplot

### boxplot
This shows the distribution of quantitative data, representing comparisons between variables. The boxplot shows the quartiles of the dataset, while the whiskers expand to show the remainder of the distribution, i.e., points indicating outliers. It is created using the boxplot() method.

```python
data = sns.load_dataset("iris")

sns.boxplot(x='species', y='sepal_width', data=data)
plt.show()
```

### Violin plot
This is similar to a boxplot, except that it provides a more advanced visualization and uses kernel density estimation to better describe the data distribution. It is created using the violinplot() method.

```python
data = sns.load_dataset("iris")

sns.violinplot(x='species', y='sepal_width', data=data)
plt.show()
```

### Strip plot
This essentially creates a scatter plot based on categories. It's created using the stripplot() method.

```python
sns.stripplot(x='species', y='sepal_width', data=data)
plt.show()
```

And there are many other types of category plots such as Factors, Swarmplot

### Distplot
Distplot is basically used for a set of univariant observations and visualizes them through a histogram i.e., only one observation is used, and hence we select one particular column from the data set.

### pairplot
A pairplot represents pairwise relationships across a data frame and supports an additional argument called hue for categorical separation. It essentially creates a combined plot between every possible numeric column, which can take some time if the data frame is very large. It is plotted using the pairplot() method.

```python
sns.pairplot(data=data, hue='species')
plt.show()
```

### Regression plot

The regression plot is mainly intended to add a visual guide that helps to emphasize the pattern in the data collection during exploration data analysis. Regression plot as the name suggests creates a regression line between two parameters and helps visualize its linear relationship.

__lmplot()__

The lmplot() method can be understood as a function that essentially plots a linear model. It creates a scatter plot with a linear fit.

```python
data = sns.load_dataset("tips")

sns.lmplot(x='total_bill', y='tip', data=data)
plt.show()
```

__MatrixPlot__

The matrix plot means plotting the matrix data where the color code diagram shows row data, column data, and value. It can be displayed using a heatmap.

```python
# loading dataset
data = sns.load_dataset("tips")

# correlation between parameters
tc = data.corr(numeric_only=True)

sns.heatmap(tc)
plt.show()
```
