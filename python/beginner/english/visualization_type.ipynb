# Types of Data Visualization

```python
import matplotlib.pyplot as plt
import pandas as pd
```

The tip database is a record of tips given by customers at a restaurant over two and a half months in the early 1990s. It contains six columns: total_bill, tip, gender, smoker, day, time, and size.

```python
data = pd.read_csv("C:/Users/LENOVO/Python/tips.csv")
data.head()
```

```python
print(data['total_bill'].max())
print(data['total_bill'].min())
print(data['total_bill'].mean())
print(data['total_bill'].median())
```

```python
data.describe().T
```

# Line plot

```python
plt.figure(figsize=(10,5))
# Line plot with days against tip
plt.plot(data['tip'], marker = 'o', linestyle='-')
# Plt.plot (data ['size'])

# Add the title to the plot
plt.title("Line Plot")

# Set the X and Y label
plt.xlabel('Days')
plt.ylabel('Tip')

plt.show()
```

# Bar chart

```python
x = data['day']
y = data['total_bill']

plt.bar(x, y)
plt.title("Tips Dataset")
plt.ylabel('Total Bill')
plt.xlabel('Day')
plt.show()
```

Available customizations for Bar Charts:
- color: For the bar face
- edgecolor: The color of the bar edge
- linewidth: The width of the bar edge
- width: The width of the bar

```python
x = data['day']
y = data['total_bill']

plt.bar(x, y, color='yellow', edgecolor='red',
		linewidth=2)

plt.title("Tips Dataset")
plt.ylabel('Total Bill')
plt.xlabel('Day')
plt.show()
```

# Histogram

__syntax:__

__matplotlib.pyplot.hist(x, bins=None, range=None, density=False, weights=None, cumulative=False, bottom=None, histtype=’bar’, align=’mid’, orientation=’vertical’, rwidth=None, log=False, color=None, label=None, stacked=False, \*, data=None, \*\*kwargs)__

```python
x = data['total_bill']

plt.hist(x)
plt.title("Tips Dataset")
plt.ylabel('Frequency')
plt.xlabel('Total Bill')
plt.show()
```

Available customizations for the histogram

- bins: The number of bins of equal width
- color: To change the face color
- edgecolor: The edge color
- linestyle: For the outline
- alpha: A mixed value, between 0 (transparent) and 1 (opaque)

```python
x = data['total_bill']

plt.hist(x, bins=25, color='blue', edgecolor='yellow',
		linestyle='-', alpha=0.595)
plt.title("Tips Dataset")
plt.ylabel('Frequency')
plt.xlabel('Total Bill')
plt.show()
```

# Scatterplot

```python
x = data['day']
y = data['total_bill']

plt.scatter(x, y)
plt.title("Tips Dataset")
plt.ylabel('Total Bill')
plt.xlabel('Day')
plt.show()
```

This graph can be more meaningful if we can add color and also change the size of the dots. We can do this using the c and s parameters of the scatter function, respectively.

The available customizations for the scatter are:

- s: marker size (can be a scalar or an array with the same size as the x or y dimension)
- c: color: the color sequence for the marker
- marker: marker style
- linewidths: marker border width
- edgecolor: marker border color
- alpha: a mixed value, between 0 (transparent) and 1 (opaque)

We can also display a color bar using the colorbar() method.

```python
x = data['day']
y = data['total_bill']

plt.scatter(x, y, c=data['size'],  s=data['total_bill'], marker='D', alpha=0.5)
plt.title("Tips Dataset")
plt.ylabel('Total Bill')
plt.xlabel('Day')
plt.colorbar()
plt.show()
```

# Pie chart

__syntax :__ \
__matplotlib.pyplot.pie(data, explode=None, labels=None, colors=None, autopct=None, shadow=False)__

```python
cars = ['AUDI', 'BMW', 'FORD',
		'TESLA', 'JAGUAR',]
data = [23, 10, 35, 15, 12]

plt.pie(data, labels=cars)
plt.title("Car data")
plt.show()
```

Available customizations for pie charts are:

- explode: Moves the plot slices.
- autopct: Labels the wedges with their numeric values.
- color: This attribute is used to color the slices.
- shadow: This attribute is used to shade the wedges.

```python
cars = ['AUDI', 'BMW', 'FORD',
		'TESLA', 'JAGUAR',]
data = [23, 13, 35, 15, 12]
explode = [0.1, 0.5, 0, 0, 0]
colors = ( "orange", "cyan", "yellow",
		"grey", "green")

plt.pie(data, labels=cars, explode=explode, autopct='%.3f%%',
		colors=colors, shadow=True)
plt.savefig("piechart.png") # If it will be stored
plt.show()
```
