### Customize

Over_value is used to find out how many types of areas and tenants exceed the threshold

```python
def over_value(x, nilai = 500):
  return sum(x > nilai)
```

```python
by_area_tenant["Rent"].agg(["min", "max", "median", "mean", over_value])
```

```python
by_area_tenant.agg({"Rent" : ["min", "max", "mean"],
                    "Size" : ["min", "max"]})
```

### Iteration

```python
for group, df in by_area_tenant:
  print(f"{group[0] and group[1]} dan jumlah barisnya {df.shape[0]}")
```

```python
for group, df in by_area_tenant:
    print(f"{group[0]} - {group[1]} | Jumlah baris: {df.shape[0]}")
```

```python
for group, df in by_area_tenant:
    print(f"{group} → rata-rata sewa: {df['Rent'].mean():.0f}")
```

## Transform

Make a median column from the rental prices

```python
house_data["median_rent_by_area"] = rent_by_area["Rent"].transform("median")
```

Only displays rows with rental prices above the median

```python
house_data[house_data["Rent"] >= house_data["median_rent_by_area"]]
```

## Average rent

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the average rent per area
avg_rent = house_data.groupby("Area Type")["Rent"].mean().sort_values()

# Plot
avg_rent.plot(kind='barh', title="Average Rent per Area")
plt.xlabel("Rent (INR)")
plt.show()
```

## Local ranking

Rank () is ranking (order) numerical given to each value in a column, usually based on the size of the value.

Note: For example 5000 appears twice → they share rankings 1 and 2 → (1+2)/2 = 1.5

```python
house_data["rank_rent_in_area"] = house_data.groupby("Area Type")["Rent"].rank(ascending=False)
# house_data ["rank_rent_in_area"] = house_data.groupby (["area type", "area locality"]) ["Rent"]. Rank (Ascending = False)
```

```python
house_data.head()
```

The house is ranked 1622 of all houses in its area, based on rental prices (from the highest to the lowest).
 
House with the top 5 rent in each type area

```python
house_data[house_data["rank_rent_in_area"] <= 5]
```
